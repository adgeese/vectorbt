```python
import multiprocessing

def merge_sort(arr, num_processes=2):
  # Base case: if the input array has fewer than 2 elements, it is already sorted
  if len(arr) < 2:
    return arr

  # Split the input array in half
  middle = len(arr) // 2
  left = arr[:middle]
  right = arr[middle:]

  # Create a new process for each half
  left_process = iprocessing.Process(target=merge_sort, args=(left, num_processes))
  right_process = iprocessing.Process(target=merge_sort, args=(right, num_processes))

  # Start the processes
  left_process.start()
  right_process.start()

  # Wait for the processes to finish
  left_process.join()
  right_process.join()

  # Merge the sorted halves and return the result
  return merge(left, right)

def merge(left, right):
  result = []
  while left and right:
    # Compare the first elements of the left and right arrays, and append the
    # smaller one to the result
    if left[0] < right[0]:
      result.append(left.pop(0))
    else:
      result.append(right.pop(0))

  # Append the remaining elements of the left or right arrays (if any) to the result
  result.extend(left)
  result.extend(right)
  return result
```

```python
sorted_list = merge_sort([5, 3, 8, 1, 9, 2], num_processes=4)
print(sorted_list)  # Output: [1, 2, 3, 5, 8, 9]
```
